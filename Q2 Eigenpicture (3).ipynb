```python
def show_eigenvectors(A):
    %display latex
    A = Matrix(ZZ, 2, 2, [2,1, 0,1])
    eigen = A.eigenvectors_right() #list eigenvectos_right() function produces
    n=len(eigen)#size of list eigen = number of eigenvectors
    evects=[eigen[i][1] for i in range (n)] #list of eigenvectors
    evals =[eigen[i][0] for i in range (n)]
    evects[0]
    Av= [A*evects[i][0] for i in range(n)]
    #same as Av=lambda*v
    #Av = [evects[i][0]*evals[i][0] for i in range(n)]
    g1 =sum([arrow2d((0,0), (evects[i][0][0], evects[i][0][1])) for i in range(n)])
    g2 = sum([arrow2d((evects[i][0][0], evects[i][0][1]), (Av[i][0]+evects[i][0][0], Av[i][1]+evects[i][0][1])) for i in range(n)])
    G=g1+g2
    return G
```
